```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
file = open('calls.csv')
df = pd.read_csv(file)
df
```

```python
G = nx.from_pandas_edgelist(df, 'caller', 'callee', ['timestamp', 'duration'], create_using=nx.DiGraph())
plt.figure(figsize=(6,6))
nx.draw(G, node_size=16)
```
